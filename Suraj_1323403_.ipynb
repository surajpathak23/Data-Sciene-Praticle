# Numpy (1 Practicals)
## Practical 1: NumPy Basics and Array Operations
- Create 1D, 2D arrays and perform indexing, slicing.
- Perform element-wise and matrix operations.
- Use functions like np.mean(), np.std(), np.dot().

```python
import numpy as np

# Arrays Setup
a1 = np.array([10, 20, 30, 40, 50])
a2 = np.array([[1, 2], [3, 4]])
b = np.array([5, 5, 5, 5, 5])

# 1. Creation, Indexing, and Slicing
print("\n1. Arrays, Indexing, Slicing:")
print("1D Array (a1):", a1)
print("2D Array (a2):\n", a2)
print("Index a1[2]:", a1[2])
print("Slice a1[1:4]:", a1[1:4])

# 2. Element-wise and Matrix Operations
print("\n2. Operations:")
print("Element-wise Sum (a1 + b):", a1 + b)
print("Matrix Product (np.dot(a2, a2)):\n", np.dot(a2, a2))

# 3. NumPy Functions
print("\n3. Functions:")
print("Mean (a1):", np.mean(a1))
print("Std Dev (a1):", np.std(a1))
```

## Practical 2: Numerical Computations and Broadcasting
- Perform broadcasting between different-shaped arrays.
- Generate random arrays and apply statistical computations.
- Implement linear algebra operations using numpy.linalg.

```python
import numpy as np

# 1. Broadcasting
print("\n1. Broadcasting:")
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([10, 20, 30]) # Will be broadcasted across A's rows
print("Array A:\n", A)
print("Broadcast Result (A + B):\n", A + B)

# 2. Random Arrays and Statistical Computations
print("\n2. Random Arrays and Statistics:")
rand_arr = np.random.randint(1, 20, size=(3, 3))
print("Random 3x3 Array:\n", rand_arr)
print("Mean:", np.mean(rand_arr))
print("Std Dev:", np.std(rand_arr))

# 3. Linear Algebra Operations (numpy.linalg)
print("\n3. Linear Algebra:")
mat = np.array([[1, 2], [3, 4]])
inv_mat = np.linalg.inv(mat)

print("Original Matrix:\n", mat)
print("Inverse Matrix (np.linalg.inv):\n", inv_mat)
print("Determinant (np.linalg.det):", np.linalg.det(mat))
```

## Practical 3: Data Cleaning and Preprocessing with Pandas
- Import dataset (e.g., CSV file).
- Handle missing values and duplicates.
- Use groupby(), describe(), and filtering techniques.


```python
import pandas as pd
# 1. Import Dataset (Creating a sample DataFrame instead of reading CSV)
df = pd.read_csv('/content/Students_data.csv')
print("\nOriginal Data:")
print(df)

print("\nCleaning Steps:")
# Missing Values
df['Score'].fillna(df['Score'].mean(), inplace=True)
# Duplicates (Dropping duplicate rows)
df.drop_duplicates(inplace=True)
print(f"Duplicates dropped. New rows: {len(df)}")

# describe()
print("Score Statistics (df.describe()):")
print(df['Score'].describe())

# groupby()
print("\nMean Score by City (groupby()):")
print(df.groupby('City')['Score'].mean())

# Filtering
filtered = df[df['Score'] > 80]
print("\nFiltered Data (Score > 80):\n", filtered)
```

## Practical 6: Data Visualization using Seaborn and Matplotlib
- Use Seaborn for heatmaps, pairplots, boxplots.
- Create correlation heatmap of dataset.
- Combine multiple Seaborn plots in one figure.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris')

print(f"\nDataset Loaded (Iris): {df.shape}")
print(df.head(2))

# --- Visualization Tasks ---
# Create a figure with subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle('Practical 6: Seaborn Visualization Summary', fontsize=16)

# 2. Create Correlation Heatmap (Seaborn)
# Calculate the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Plot the Heatmap on the first subplot
sns.heatmap(
    corr_matrix,
    annot=True,          # Show correlation values on the map
    cmap='coolwarm',     # Color map
    linewidths=0.5,      # Line separation
    fmt=".2f",           # Format to 2 decimal places
    cbar_kws={'shrink': 0.8}, # Shrink color bar
    ax=axes[0]           # Assign to the left subplot
)
axes[0].set_title('Correlation Heatmap')

# 3. Create Boxplot (Comparison of distributions)
# Plot Boxplot on the second subplot
sns.boxplot(
    x='species',         # Categorical variable
    y='petal_length',    # Numerical variable
    data=df,
    palette='viridis',   # Color palette
    ax=axes[1]           # Assign to the right subplot
)
axes[1].set_title('Boxplot: Petal Length by Species')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Petal Length (cm)')


# Display the combined plots
plt.tight_layout(rect=[0, 0.03, 1, 0.9])
plt.show()

# --- Extra (For reference, if needed) ---
# For Pairplots, run separately as they require a lot of space:
print("\nPairplot (Displayed in a separate window):")
sns.pairplot(df, hue='species')
plt.show()
```

## Practical 7: Data Preprocessing and Feature Engineering
- Handle missing data using SimpleImputer.
- Encode categorical variables using LabelEncoder or OneHotEncoder.
- Apply feature scaling using StandardScaler or MinMaxScaler.
- Split the dataset into training and testing sets using train_test_split.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Sample Data (Includes NaN for imputation, string for encoding)
data = {
    'Age': [30, 40, np.nan, 25, 50],
    'Salary': [50, 60, 70, np.nan, 80],
    'City': ['Paris', 'Tokyo', 'Paris', 'Berlin', 'Tokyo'],
    'Target': [0, 1, 0, 1, 1]
}
df = pd.DataFrame(data)

# Separate X and y
X = df.drop('Target', axis=1)
y = df['Target']

# Define column types
num_cols = ['Age', 'Salary']
cat_cols = ['City']

# 2. Define Preprocessing Steps (Pipeline)
preprocessor = ColumnTransformer(
    transformers=[
        # Numerical Pipeline: Impute (Mean) -> Scale (StandardScaler)
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ]), num_cols),

        # Categorical Pipeline: Encode (OneHotEncoder)
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])

# Apply preprocessing to features
X_processed = preprocessor.fit_transform(X)

print("\nFeatures Processed (Imputed, Scaled, Encoded):")
print(X_processed)
print(f"Processed Shape: {X_processed.shape}")

# 3. Split the Dataset (train_test_split)
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.3, random_state=42
)

print("\nDataset Split (30% Test):")
print(f"X_train size: {X_train.shape}")
print(f"X_test size: {X_test.shape}")
```

## Practical 8: Supervised Learning – Regression Models
- Load a dataset (e.g., Student Score or House Price Prediction).
- Train Linear Regression and Polynomial Regression models.
- Predict outcomes and evaluate with metrics: MAE, MSE, RMSE, and R² score.
- Visualize regression line and residual errors using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# 1. Load/Create Dataset (Dummy Data for simplicity)
np.random.seed(42)
X = np.arange(10).reshape(-1, 1) # Features (1 to 10)
y = 2 * X.flatten() + 5 + np.random.randn(10) * 3 # Target (linear relationship + noise)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nData Split: Train size {X_train.shape[0]}, Test size {X_test.shape[0]}")

# --- Linear Regression ---
# 2. Train Model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# 3. Predict and Evaluate
y_pred_lin = lin_reg.predict(X_test)
print("\n--- Linear Regression Metrics ---")
print(f"MAE: {mean_absolute_error(y_test, y_pred_lin):.2f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_lin):.2f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_lin)):.2f}")
print(f"R^2 Score: {r2_score(y_test, y_pred_lin):.2f}")

# --- Polynomial Regression (Degree 2) ---
# Transform features
poly_features = PolynomialFeatures(degree=2)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.transform(X_test)

# 2. Train Model
poly_reg = LinearRegression()
poly_reg.fit(X_poly_train, y_train)

# 3. Predict and Evaluate
y_pred_poly = poly_reg.predict(X_poly_test)
print("\n--- Polynomial Regression (Deg 2) Metrics ---")
print(f"R^2 Score: {r2_score(y_test, y_pred_poly):.2f}")


# 4. Visualize Regression Line and Residual Errors
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Practical 8: Regression Analysis', fontsize=16)

# Plot 1: Regression Line (Linear Model)
axes[0].scatter(X_test, y_test, color='blue', label='Actual Data')
axes[0].plot(X_test, y_pred_lin, color='red', linewidth=2, label='Linear Regression Line')
axes[0].set_title('Linear Regression Line')
axes[0].legend()

# Plot 2: Residual Errors (Linear Model)
residuals = y_test - y_pred_lin
axes[1].scatter(y_pred_lin, residuals, color='green')
axes[1].hlines(y=0, xmin=y_pred_lin.min(), xmax=y_pred_lin.max(), color='red', linestyle='--')
axes[1].set_title('Residual Errors')
axes[1].set_xlabel('Predicted Values')
axes[1].set_ylabel('Residuals (Actual - Predicted)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Practical 9: Supervised Learning – Classification Models
- Load dataset (e.g., Iris, Breast Cancer, or custom dataset).
- Train models: Logistic Regression, K-Nearest Neighbors (KNN), and Decision Tree Classifier.
- Evaluate models using Confusion Matrix, Accuracy, Precision, Recall, and F1-score.
- Visualize decision boundaries and confusion matrix heatmaps.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.datasets import load_iris


# 1. Load Dataset (Iris)
iris = load_iris(as_frame=True)
X = iris.data
y = iris.target
target_names = iris.target_names

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nData Split: Train size {X_train.shape[0]}, Test size {X_test.shape[0]}")

# Define the models
models = {
    "Logistic Regression": LogisticRegression(max_iter=200, random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

# --- Train, Predict, and Evaluate ---
print("\n--- Model Evaluation ---")
best_model = None
best_accuracy = 0

for name, model in models.items():
    # 2. Train Model
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # 3. Evaluate Models
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    print(f"\nModel: {name}")
    print(f"Accuracy: {accuracy:.4f} | Precision: {precision:.4f} | Recall: {recall:.4f} | F1-Score: {f1:.4f}")

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model

# --- Visualization (Confusion Matrix for the Best Model) ---
y_pred_best = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(7, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=target_names,
    yticklabels=target_names
)
plt.title(f'Confusion Matrix Heatmap: {best_model.__class__.__name__}')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
